```python
df["age"] = df["age"].fillna(df["age"].mean())
```

```python
df["age"]=df["age"].astype(int)
```

```python
def categorize_age(age):
    if age < 12:
        return 'Child'
    elif age < 18:
        return 'Teen'
    elif age < 35:
        return 'Young Adult'
    elif age < 60:
        return 'Adult'
    else:
        return 'Senior'
df['Age Group'] = df['age'].apply(categorize_age)
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt  # Also needed for plotting

# Step 1: Load data
jk = pd.read_csv('products.csv')

# Clean column names
jk.columns = jk.columns.str.strip().str.lower()
print("âœ… Columns loaded:", jk.columns.tolist())

# Step 2: Derive margin percentage if missing
if 'cost_price' in jk.columns and 'unit_price' in jk.columns:
    jk['margin_amount'] = jk['unit_price'] - jk['cost_price']
    jk['calculated_margin_pct'] = (jk['margin_amount'] / jk['unit_price']) * 100
    if 'margin_pct' not in jk.columns:
        jk['margin_pct'] = jk['calculated_margin_pct']
else:
    raise KeyError("Missing required columns: 'cost_price' or 'unit_price'.")

# Step 3: Summary statistics
print("\nðŸ“Š Summary statistics:")
print(jk[['cost_price', 'unit_price', 'margin_pct', 'margin_amount']].describe())

# Step 4: Top products by margin and price
top_by_price = jk.sort_values(by='unit_price', ascending=False).head(10)
top_by_margin = jk.sort_values(by='margin_pct', ascending=False).head(10)

print("\nðŸ· Top 10 Products by Unit Price:")
print(top_by_price[['product_name',Â 'unit_price']])
```
