```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("customers.csv")
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["age"] = df["age"].fillna(df["age"].mean())
```

```python
df["age"]=df["age"].astype(int)
```

```python
def categorize_age(age):
    if age < 12:
        return 'Child'
    elif age < 18:
        return 'Teen'
    elif age < 35:
        return 'Young Adult'
    elif age < 60:
        return 'Adult'
    else:
        return 'Senior'
df['Age Group'] = df['age'].apply(categorize_age)
df
```

```python
df.to_csv("C:\\Users\\vvams\\OneDrive\\Desktop\\Cleaned Data Customers.csv",index=False)
```
